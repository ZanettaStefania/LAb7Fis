```python
# extract in the directory
def extract_features(directory):
  df_list=[]
  for name in os.listdir(directory):
    print(name)
    if(name != ".ipynb_checkpoints"):
      # Load an image from file
      filename = directory + '/' + name
      X = pd.read_csv(filename)
      df_list.append(X)
  return df_list
```

## STACKED BARCHART DATA
Select 3 tree types & take all the states where all 3 types of tree can be found

Select only the cities where 'Platanus acerifolia', 'Lagerstroemia indica' and 'Acer platanoides' can be found.

```python
Y_city_sciencename=pd.DataFrame(data[['city', 'scientific_name']])
Y_city_sciencename=Y_city_sciencename.dropna()
```

```python
Y_city_sciencename_count=Y_city_sciencename.groupby(['city','scientific_name'])['scientific_name'].count().reset_index(name="Total")
```

```python
Tree3=Y_city_sciencename_count[(Y_city_sciencename_count['scientific_name'] == 'Platanus acerifolia') | ( Y_city_sciencename_count['scientific_name'] == 'Lagerstroemia indica') | ( Y_city_sciencename_count['scientific_name'] == 'Acer platanoides') ]
Tree3 = Tree3.groupby('city')['city'].count().reset_index(name="Total")
Tree3=Tree3[Tree3['Total'] == 3]
Tree3
```

```python
Platanus=Y_city_sciencename_count[(Y_city_sciencename_count['scientific_name'] == 'Platanus acerifolia')]
Platanus=Platanus.rename(columns={'Total': 'Platanus_Acerifolia'})
Platanus=Platanus.drop(columns=['scientific_name'])
Platanus = Platanus[Platanus.city.isin(Tree3.city)]
Platanus
```

```python
Acer=Y_city_sciencename_count[(Y_city_sciencename_count['scientific_name'] == 'Acer platanoides')]
Acer=Acer.rename(columns={'Total': 'Acer_Platanoides'})
Acer=Acer.drop(columns=['scientific_name'])
Acer = Acer[Acer.city.isin(Tree3.city)]
Acer
```

```python
Lagerstroemia=Y_city_sciencename_count[(Y_city_sciencename_count['scientific_name'] == 'Lagerstroemia indica')]
Lagerstroemia=Lagerstroemia.rename(columns={'Total': 'Lagerstroemia_Indica'})
Lagerstroemia=Lagerstroemia.drop(columns=['scientific_name'])
Lagerstroemia = Lagerstroemia[Lagerstroemia.city.isin(Tree3.city)]
Lagerstroemia
```

Count all the other trees that can be found in the selected cities.

```python
Other=Y_city_sciencename_count[(Y_city_sciencename_count['scientific_name'] != 'Platanus acerifolia') | ( Y_city_sciencename_count['scientific_name'] != 'Lagerstroemia indica') | ( Y_city_sciencename_count['scientific_name'] != 'Acer platanoides') ]
Other = Other.groupby('city')['city'].count().reset_index(name="Total")
Other=Other.rename(columns={'Total': 'Other'})
Other = Other[Other.city.isin(Tree3.city)]
Other
```

Merge all the tables together (1 raw for each city)

```python
result = pd.merge(Acer, Lagerstroemia, on=["city"])
result = pd.merge(result, Platanus, on=["city"])
result = pd.merge(result, Other, on=["city"])
result
```

```python
result.to_csv("Top15_city_3Tree_others.csv", index=False)
```

## HEATMAP DATA
Select 3 tree types & take the top 5 / top 10 / top 15 states where all 3 types of tree can be found and sort them in descending order

```python
Y_city_sciencename=pd.DataFrame(data[['city', 'scientific_name']])
Y_city_sciencename=Y_city_sciencename.dropna()
```

```python
Y_city_sciencename_count=Y_city_sciencename.groupby(['city','scientific_name'])['scientific_name'].count().reset_index(name="Total")
```

Select only the cities where 'Platanus acerifolia', 'Lagerstroemia indica' and 'Acer platanoides' can be found, with their tree count.

```python
AllTreeTypes_withcount=Y_city_sciencename_count[(Y_city_sciencename_count['scientific_name'] == 'Platanus acerifolia') | ( Y_city_sciencename_count['scientific_name'] == 'Lagerstroemia indica') | ( Y_city_sciencename_count['scientific_name'] == 'Acer platanoides') ]
Platanus_city_sciencenam = AllTreeTypes_withcount.groupby('city')['city'].count().reset_index(name="Total")
Tree3=Platanus_city_sciencenam[Platanus_city_sciencenam['Total'] == 3]
Tree3_withcount= AllTreeTypes_withcount[AllTreeTypes_withcount.city.isin(Tree3.city)]
Tree3_withcount
```

Count all the other trees that can be found in the selected cities.

```python
Other=Y_city_sciencename_count[(Y_city_sciencename_count['scientific_name'] != 'Platanus acerifolia') | ( Y_city_sciencename_count['scientific_name'] != 'Lagerstroemia indica') | ( Y_city_sciencename_count['scientific_name'] != 'Acer platanoides') ]
Other = Other.groupby('city')['city'].count().reset_index(name="Total")
Other['scientific_name'] = 'Other'
Other = Other[Other.city.isin(Tree3.city)]
Other
```

Sum all the other trees that can be found in the selected cities.

```python
temp = pd.concat([Tree3_withcount, Other])
SumOfTreeType = pd.DataFrame(temp.groupby('city')['Total'].sum().reset_index(name="SumOfTreeType"))
city_Sum_Total = pd.merge(SumOfTreeType, temp, on=["city"])
city_Sum_Total = city_Sum_Total.sort_values(by=['SumOfTreeType'], ascending=False)
city_Sum_Total
```

After sorting the cities by their respective sums select the top 5.

```python
Sorted5City_Total=city_Sum_Total.head(20)
Sorted5City_Total=Sorted5City_Total.drop(columns=['SumOfTreeType'])
Sorted5City_Total
```

```python
Sorted5City_Total.to_csv("Top5_3Tree&others.csv", index=False)
```
